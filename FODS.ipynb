```python

import pandas as pd
import numpy as np

# 1. Create a Chess competition dataset
data = {
    'Player_ID': [1, 2, 3, 4, 5],
    'Player_Name': ['Jana', 'Hari', 'Thanush', 'Divya', 'Ramya'],
    'Rating': [2200, 2100, 2300, 2150, 2250],
    'Games_Played': [10, 12, 8, 15, 11],
    'Games_Won': [6, 8, 5, 9, 7],
    'Games_Drawn': [2, 1, 2, 3, 2],
    'Games_Lost': [2, 3, 1, 3, 2],
    'Total_Score': [20, 19, 17, 24, 21]
}

# Create DataFrame
df = pd.DataFrame(data)

# 2. Identify the features in the dataset
print("Features in the dataset:")
features = df.columns
print(features)

# 3. Identify the datatype of each feature
print("\nData types of each feature:")
data_types = df.dtypes
print(data_types)

# 4. Check the five-point summary of the features
print("\nFive-point summary of numerical features:")
numerical_summary = df.describe(include=[np.number])
print(numerical_summary)

print("\nFive-point summary of categorical features:")
categorical_summary = df.describe(include=[object])
print(categorical_summary)

# 5. Explanation
print("\nExplanation:")
print("The dataset contains information about players in a chess competition.")
print("Numerical features include 'Rating', 'Games_Played', 'Games_Won', 'Games_Drawn', 'Games_Lost', and 'Total_Score'.")
print("Categorical features include 'Player_ID' and 'Player_Name'.")
print("The five-point summary for numerical features gives us an idea of the distribution of ratings, games played, wins, draws, losses, and scores.")
print("The categorical summary provides frequency information for non-numeric attributes.")
print("Actions to consider as a data scientist:")
print("- Data Validation: Ensure the data is accurate and complete.")
print("- Feature Engineering: Compute additional features such as win rate or average score if necessary.")
print("- Exploratory Data Analysis (EDA): Analyze relationships between features, such as how player rating correlates with performance metrics.")
print("- Modeling: Build models to predict outcomes, such as predicting future scores based on past performance.")
```

```python
# Checking for missing values
missing_values = df.isnull().sum()
print("Missing values in each feature:")
print(missing_values)
```

```python
# Adding a new column for win rate
df['Win_Rate'] = df['Games_Won'] / df['Games_Played']
print("\nWin Rate of each player:")
print(df[['Player_Name', 'Win_Rate']])
```

```python
# Adding a new column for loss rate
df['Loss_Rate'] = df['Games_Lost'] / df['Games_Played']
print("\nLoss Rate of each player:")
print(df[['Player_Name', 'Loss_Rate']])
```

```python

# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Display the correlation matrix
print("\nCorrelation matrix of numerical features:")
print(correlation_matrix)
```

```python
# Sorting players by Total_Score
top_performers = df.sort_values(by='Total_Score', ascending=False)
print("\nTop performers based on Total Score:")
print(top_performers[['Player_Name', 'Total_Score']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the distribution of player ratings
plt.figure(figsize=(8, 4))
sns.histplot(df['Rating'], kde=True, bins=5)
plt.title('Distribution of Player Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of win rate by player
plt.figure(figsize=(8, 4))
sns.barplot(x='Player_Name', y='Win_Rate', data=df, palette='viridis')
plt.title('Win Rate by Player')
plt.xlabel('Player Name')
plt.ylabel('Win Rate')
plt.show()
```

```python
# Scatter plot to show the relationship between Rating and Total_Score
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Rating', y='Total_Score', data=df, s=100, color='blue')
plt.title('Rating vs. Total Score')
plt.xlabel('Rating')
plt.ylabel('Total Score')
plt.show()
```

```python
# Plotting distribution of game outcomes
outcomes = df[['Games_Won', 'Games_Drawn', 'Games_Lost']].sum()
outcomes.plot(kind='pie', autopct='%1.1f%%', startangle=90, labels=['Won', 'Drawn', 'Lost'], colors=['green', 'orange', 'red'])
plt.title('Distribution of Game Outcomes')
plt.ylabel('')
plt.show()
```

```python
# Box plot of total scores
plt.figure(figsize=(8, 4))
sns.boxplot(y='Total_Score', data=df, palette='coolwarm')
plt.title('Box Plot of Total Scores')
plt.ylabel('Total Score')
plt.show()
```
